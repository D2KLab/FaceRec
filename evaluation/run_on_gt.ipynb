```python
tr = Tracker(project='memad_gt')
```

```python
from tqdm.notebook import tqdm

all_results = []
for i, x in tqdm(df.iterrows(), total=len(df)):
    start = int(x['start'])
    end = int(x['end'])
    locator = x['locator']
    v = limecraft.locator2video(locator)
    print(v)
    res = tr.run(v, export_frames=True, fragment=f'{start-1},{end}', video_id=x['media'], cluster_features=False)
    print(res)
    all_results.append(res)
```

```python
import json
with open('results_memad_gt2.json', 'w') as f:
    json.dump(all_results, f)
```

```python
from src import clusterize

clusters  = []
for r in all_results:
    c = clusterize.main(clusterize.from_dict(r), confidence_threshold=0.0, merge_cluster=True, min_length=1)
    clusters.append(c)
    
with open('results_memad_gt_clusters2.json', 'w') as f:
    json.dump(clusters, f)
```
